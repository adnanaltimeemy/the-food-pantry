```python
import pandas as pd
import numpy as np
```

```python
businesses = [
    {"name":"Syria Sweets Ltd","city":"Manchester","type":"Dessert","avg_price":12},
    {"name":"Mr. Knafah","city":"Manchester","type":"Dessert","avg_price":10},
    {"name":"Al Hamra","city":"Manchester","type":"Candy Store","avg_price":8},
    {"name":"Emesa sweet & dessert","city":"Manchester","type":"Dessert","avg_price":9},
    {"name":"Sweetland London","city":"London","type":"Candy Store","avg_price":10},
    {"name":"I baklawa LTD","city":"London","type":"Baklava & Sweets","avg_price":11}
]

df_business = pd.DataFrame(businesses)
df_business
```

```python
menu = [
    {"shop":"Syria Sweets Ltd","item":"Konafa","price":10},
    {"shop":"Syria Sweets Ltd","item":"Basbousa","price":8},
    {"shop":"Mr. Knafah","item":"Kunafa","price":12},
    {"shop":"Mr. Knafah","item":"Baklava","price":7},
    {"shop":"Al Hamra","item":"Baklava Box","price":8},
    {"shop":"I baklawa LTD","item":"Assorted Baklava","price":11},
]

df_menu = pd.DataFrame(menu)
df_menu
```

```python
def estimate_customers(city):
    if city == "Manchester":
        return np.random.randint(100, 250)  # more local traffic
    else:
        return np.random.randint(80, 200)
```

```python
import random

results = []
for idx,row in df_business.iterrows():
    customers = estimate_customers(row['city'])
    avg_spend = row['avg_price']
    weekly_revenue = customers * avg_spend
    cogs = weekly_revenue * 0.4
    weekly_profit = weekly_revenue - cogs
    yearly_profit = weekly_profit * 50  # 50 working/peak weeks
    results.append({
        "shop":row['name'],
        "weekly_profit":weekly_profit,
        "yearly_profit":yearly_profit
    })

df_profit = pd.DataFrame(results)
df_profit
```

```python
df_profit = df_profit.merge(df_business, left_on="shop", right_on="name")
df_profit.groupby("city")[["weekly_profit","yearly_profit"]].sum()
```
