```python
# Step 1: Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Show plots in notebook
%matplotlib inline

print("Libraries imported successfully")
```

```python
# Step 2: Create business info dataset

business_data = {
    "Business Name": [
        "Barbakan Delicatessen",
        "Salvi's Manchester"
    ],
    "Location": [
        "Chorlton, Manchester",
        "Corn Exchange, Manchester"
    ],
    "Business Type": [
        "Polish Bakery & Delicatessen",
        "Italian Restaurant & Deli"
    ],
    "Years Active": [
        60,
        15
    ]
}

business_df = pd.DataFrame(business_data)

business_df
```

```python
# Step 3: Create synthetic monthly sales data

months = [
    "Jan", "Feb", "Mar", "Apr", "May", "Jun",
    "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"
]

np.random.seed(42)

barbakan_sales = np.random.randint(35000, 55000, size=12)
salvis_sales = np.random.randint(50000, 80000, size=12)

sales_df = pd.DataFrame({
    "Month": months,
    "Barbakan Sales (£)": barbakan_sales,
    "Salvis Sales (£)": salvis_sales
})

sales_df
```

```python
# Step 4: Calculate profit margins

barbakan_margin = 0.18
salvis_margin = 0.22

sales_df["Barbakan Profit (£)"] = sales_df["Barbakan Sales (£)"] * barbakan_margin
sales_df["Salvis Profit (£)"] = sales_df["Salvis Sales (£)"] * salvis_margin

sales_df
```

```python
# Step 5: Customer demographics dataset

demographics_data = {
    "Age Group": [
        "18-24",
        "25-34",
        "35-44",
        "45-54",
        "55-64",
        "65+"
    ],
    "Barbakan Customers (%)": [
        10,
        20,
        25,
        20,
        15,
        10
    ],
    "Salvis Customers (%)": [
        18,
        35,
        22,
        12,
        8,
        5
    ]
}

demographics_df = pd.DataFrame(demographics_data)

demographics_df
```

```python
# Step 6: Annual totals

barbakan_total_sales = sales_df["Barbakan Sales (£)"].sum()
salvis_total_sales = sales_df["Salvis Sales (£)"].sum()

barbakan_total_profit = sales_df["Barbakan Profit (£)"].sum()
salvis_total_profit = sales_df["Salvis Profit (£)"].sum()

summary_df = pd.DataFrame({
    "Business": ["Barbakan", "Salvi's"],
    "Annual Sales (£)": [barbakan_total_sales, salvis_total_sales],
    "Annual Profit (£)": [barbakan_total_profit, salvis_total_profit]
})

summary_df
```

```python
# Step 7: Plot sales

plt.figure()

plt.plot(sales_df["Month"], sales_df["Barbakan Sales (£)"])
plt.plot(sales_df["Month"], sales_df["Salvis Sales (£)"])

plt.title("Monthly Sales Comparison")
plt.xlabel("Month")
plt.ylabel("Sales (£)")

plt.show()
```

```python
# Step 8: Profit comparison bar chart

plt.figure()

plt.bar(summary_df["Business"], summary_df["Annual Profit (£)"])

plt.title("Annual Profit Comparison")
plt.xlabel("Business")
plt.ylabel("Profit (£)")

plt.show()
```

```python
# Step 9: Demographics comparison

plt.figure()

plt.plot(
    demographics_df["Age Group"],
    demographics_df["Barbakan Customers (%)"]
)

plt.plot(
    demographics_df["Age Group"],
    demographics_df["Salvis Customers (%)"]
)

plt.title("Customer Demographics by Age")
plt.xlabel("Age Group")
plt.ylabel("Percentage")

plt.show()
```

```python
# Step 10: Save files

sales_df.to_csv("sales_data.csv", index=False)
demographics_df.to_csv("demographics_data.csv", index=False)
summary_df.to_csv("summary_data.csv", index=False)

print("Datasets saved successfully")
```
